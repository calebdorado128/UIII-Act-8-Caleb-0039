## Realizando simulacion de inteligencia artificial con if, if else, elif y while tipo menu
# Problema 15: Cocinero, sarten

```python
# --- SIMULACIÓN DE IA: EL SARTÉN INTELIGENTE DEL CHEF ---

# Configuración inicial (Estado de la cocina)
temperatura = 0
ingrediente = "Ninguno"
estado_comida = "Vacío"
energia_estufa = 100
cocina_encendida = "s"

print("IA: ¡Hola, Chef! Soy tu Sartén Inteligente 3000. He detectado tu presencia.")
print("IA: Mis sensores están listos para cocinar algo delicioso.")

# Ciclo principal para mantener el programa interactivo
while cocina_encendida == "s" or cocina_encendida == "S":
    print("\n--- MENÚ DE CONTROL DEL CHEF ---")
    print(f"Estado Actual: {estado_comida} | Temp: {temperatura}°C | Gas: {energia_estufa}%")
    print("1. Encender fuego (Calentar)")
    print("2. Añadir ingrediente (Huevo, Carne o Vegetales)")
    print("3. Iniciar cocción (Cocinar)")
    print("4. Sensor de seguridad (Enfriar)")
    print("5. Servir plato y Salir")

    opcion = input("CHEF (Elige 1-5): ")

    # Lógica de la IA utilizando condicionales
    if opcion == "1":
        # Operador aritmético (+) y relacional (<=)
        if energia_estufa > 0:
            temperatura = temperatura + 40
            energia_estufa = energia_estufa - 5
            print(f"IA: Aumentando el fuego. La temperatura ahora es de {temperatura}°C.")
            # IA advirtiendo sobre peligro usando if simple
            if temperatura > 150:
                print("IA: ¡PELIGRO! El sartén está demasiado caliente. ¡Cuidado con el aceite!")
        else:
            print("IA: Error. No queda gas en la estufa.")

    elif opcion == "2":
        print("IA: ¿Qué deseas cocinar? (1. Huevo | 2. Carne | 3. Vegetales)")
        tipo = input("CHEF (Elige 1-3): ")

        if tipo == "1":
            ingrediente = "Huevo"
            estado_comida = "Crudo"
            print("IA: Has añadido un Huevo. ¡Cuidado que no caigan cáscaras!")
        elif tipo == "2":
            ingrediente = "Carne"
            estado_comida = "Cruda"
            print("IA: Carne añadida. El sensor detecta que es de buena calidad.")
        elif tipo == "3":
            ingrediente = "Vegetales"
            estado_comida = "Crudos"
            print("IA: Vegetales añadidos. ¡Muy saludable, Chef!")
        else:
            print("IA: No reconozco ese ingrediente. El sartén sigue vacío.")

    elif opcion == "3":
        # Uso de operadores lógicos (and) y relacionales (>=)
        if ingrediente != "Ninguno" and temperatura >= 80 and temperatura <= 150:
            estado_comida = "Perfectamente Cocinado"
            print(f"IA: ¡Excelente técnica! El {ingrediente} se ha cocinado a {temperatura}°C.")
        elif ingrediente == "Ninguno":
            print("IA: Chef, no puedo cocinar el aire. Por favor, añade un ingrediente primero.")
        elif temperatura < 80:
            print("IA: El sartén está muy frío. Necesito más fuego para empezar.")
        else:
            estado_comida = "Quemado"
            print("IA: ¡OH NO! Olvidaste el sartén en el fuego. La comida se ha quemado.")

    elif opcion == "4":
        # Operador aritmético (-)
        if temperatura > 0:
            temperatura = temperatura - 30
            print(f"IA: Aplicando ventilación. Temperatura reducida a {temperatura}°C.")
        else:
            print("IA: El sartén ya está a temperatura ambiente.")

    elif opcion == "5":
        if estado_comida == "Perfectamente Cocinado":
            print(f"IA: ¡Felicidades Chef! Has servido un {ingrediente} delicioso.")
        elif estado_comida == "Quemado":
            print("IA: Serviste comida quemada... El cliente no parece muy feliz.")
        else:
            print("IA: El plato está vacío o la comida sigue cruda. Sesión terminada.")

        cocina_encendida = "n" # Termina el ciclo while

    else:
        print("IA: Chef, esa opción no está en mi panel de control. Intenta del 1 al 5.")

    # Control de continuidad
    if cocina_encendida != "n":
        cocina_encendida = input("\nIA: ¿Quieres seguir cocinando? (s/n): ")

print("\nIA: Sistema apagado. ¡Gracias por cocinar conmigo, Chef!")
```
